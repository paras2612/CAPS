```python
from cdlib import algorithms,viz
import networkx as nx
from natsort import natsorted # pip install natsort
import os,re,glob
import pandas as pd
from cdlib import TemporalClustering
from networkx.generators.community import LFR_benchmark_graph
from cdlib import readwrite
import csv



path = r'C:\Users\dvdoner\Desktop\Causality\New_CGR'
```

```python
tc = TemporalClustering()
count=0
com_id=0
lst_values=[]
all_data_rows=[]
frames=[]
all_communities=[]
community_names=[]
tc = TemporalClustering()
for filename in natsorted(os.listdir(path)):
    if filename.endswith(".csv"):
        #print(os.path.join(path, filename))

        csv_path=r'C:\Users\dvdoner\Desktop\Causality\New_CGR\\'+filename

        df = pd.read_csv(csv_path)
        G = nx.from_pandas_edgelist(df, "from", "to",create_using=nx.DiGraph())

        a=nx.is_directed_acyclic_graph(G)
        
        if a==True:
          coms = algorithms.siblinarity_antichain(G,Lambda=1)
          #readwrite.write_community_csv(coms, "communities.csv", ",")
          pos = nx.circular_layout(G)
          #viz.plot_network_clusters(G, coms, pos, plot_overlaps=False,  figsize=(9, 9))


          tc.add_clustering(coms, filename)


          no_edges=G.number_of_edges()
          no_nodes=G.number_of_nodes()
          no_communities=len(coms.communities)
          count+=1
          lst_values.append(no_edges)
          lst_values.append(no_nodes)
          lst_values.append(no_communities)

          
          #print("{}-# of Edges: {}\n# of Nodes: {}\n# of communities {}".format(count,no_edges,no_nodes,no_communities))
          all_data_rows.append(lst_values)
          lst_values=[]        




          all_communities.append(coms.communities)

          emp_str = ""
          for m in filename:
              if m.isdigit():
                  emp_str = emp_str + m
            
          community_names.append(emp_str)

          
         
          com_id+=1

                 
```
